# Diabetes Dataset linear regression

## Dataset from SKLEARN

```python
from matplotlib import pyplot as pt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
import pandas as pd
```

```python
df=datasets.load_breast_cancer()
df
```

```python
df
x=pd.DataFrame(df.data,columns=df.feature_names)

```

```python
y=pd.Series(df.target)
```

```python
df.feature_names
```

```python
model =LinearRegression()
trainx,predx,trainy,actualy=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
trainx
```

```python


# Now fit the model
model.fit(trainx, trainy)

```

```python
prediction=model.predict(predx)
```

```python
msa=mean_squared_error(actualy,prediction)
print('mean_squared_error is ',msa)
mean_absolute_error(actualy,prediction)
```
