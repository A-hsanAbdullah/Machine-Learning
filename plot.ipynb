```python
from matplotlib import pyplot as pplot
import random
```

```python
Y=[(random.randint(1,100)) for i in range(10)]
X=[i+1 for i in range(10)]
Y1=[(random.randint(1,100)) for i in range(10)]
X2=[i+1 for i in range(10)]
pplot.subplot(2,1,1)
pplot.scatter(X,Y,marker='*',c='r',s=100)
pplot.subplot(2,1,2)
pplot.scatter(X2,Y1,marker='.',c='g',s=100)
pplot.show()
```
