# Analysis Of teams in La Liga cup 

***
## *Reading and cleaning the dataset*
***

```python
import pandas as pd
```

```python
data = pd.read_csv("Laliga.csv",delimiter=",")
```

```python
data.columns=data.iloc[0]
```

```python
data=data[1:]
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data=data.replace("-",0)
```

```python
data["Debut"]
```

***
### **All the teams which have started playing between 1930-1980**
***

```python
index=0
for i in data["Debut"]:
   if '-' in str(i) :
      parts=i.split("-")
      data.loc[index,"Debut"]=int(parts[0])
   else:
      data.loc[index,"Debut"]=int(i)
      
   index+=1
updated_data=data[(data["Debut"]>=1930) & (data["Debut"]<=1980)]
updated_data.reset_index(drop=True, inplace=True)
updated_data=pd.DataFrame(updated_data)
print(updated_data["Team"])
```

***
### **Teams which came Top 5 in terms of points**
*** 

```python
data['Points']=data['Points'].astype(int)
data['Points']=sorted(data['Points'],reverse=True)
    
```

```python
data[0:5]
```

***
### **Team which has maximum and minimum goal difference**
***

```python

def Goal_diff_count(data):
    data["GoalsAgainst"]=data["GoalsAgainst"].astype(int)
    data["GoalsFor"]=data["GoalsFor"].astype(int)
    data["Goals_Diff"]=data["GoalsFor"]-data["GoalsAgainst"]
Goal_diff_count(data)
data
```

```python
dt=data.sort_values(by="Goals_Diff",ascending=False)
dt[(dt['Goals_Diff']==dt['Goals_Diff'].max()) | (dt['Goals_Diff']==dt['Goals_Diff'].min())]
```

***
### **5 teams which has the highest Winning percentage**
***

```python
data['GamesWon']=data['GamesWon'].astype(int)
data["GamesPlayed"]=data["GamesPlayed"].astype(int)
data['Wining_Percent']=(data['GamesWon'] /data["GamesPlayed"])*100
data
```

```python
data=data.fillna(0)
```

```python
W_data=data.sort_values(by="Wining_Percent",ascending=False)
W_data
```

***
### **Group teams based on their “Best position” and print the sum of their points for all positions**
***

```python
data['BestPosition']=data['BestPosition'].astype(int)
data['Points']=data['Points'].astype(int)
Best_data=data.sort_values(by="BestPosition")
pos=1
points=0
p_list=[]
for i, j in zip(Best_data['BestPosition'], Best_data['Points']):
    if i == pos:
        points += j
    else:
        # Handle skipped positions
        while pos < i:
            p_list.append(points)
            points = 0
            pos += 1
        points += j

p_list.append(points)
print("*"*23,"\nBest Position\tPoints")
for i in range(len(p_list)): 
    print(i+1,"\t\t",p_list[i])
print("*"*23)

        
```
